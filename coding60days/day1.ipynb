### [704. Binary Search](https://leetcode.com/problems/binary-search/)
Personally, I prefer using open interval for general case.

```python
# open interval
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        l, r = -1, len(nums)

        while l + 1 < r:
            m = (l + r) // 2
            if nums[m] < target:
                l = m
            else:
                r = m
        
        return r if 0 <= r < len(nums) and nums[r] == target else -1

class Solution:
    def search(self, nums: List[int], target: int) -> int:
        l, r = -1, len(nums)

        while l + 1 < r:
            m = (l + r) // 2
            if nums[m] < target:
                l = m
            elif nums[m] > target:
                r = m
            else:
                return m
        
        return -1
```

### [27. Remove Element](https://leetcode.com/problems/remove-element/)

```python
# brute force
# suggested by programmercarl.com
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        i, l = 0, len(nums)
        while i < l:
            if nums[i] == val: # 找到等于目标值的节点
                for j in range(i+1, l): # 移除该元素，并将后面元素向前平移
                    nums[j - 1] = nums[j]
                # 将长度减1
                l -= 1
                # 重复检查当前位置 i，因为现在这个位置的元素可能还是等于目标值
                i -= 1
            i += 1
        return l

# I think the key point here is shifting elements to the left when removing an element and keeping track of the length of the array.
```

```python
# fast and slow pointers
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        slow = 0
        for fast in range(len(nums)):
            if nums[fast] != val:
                # If the current element is not equal to the target value,
                # we copy it to the slow pointer's position.
                nums[slow] = nums[fast]
                # Then we increment the slow pointer.
                slow += 1
        return slow

# The key point here is that we only increment the slow pointer when we find an element that is not equal to the target value.
# This way, we effectively overwrite the elements that are equal to the target value with the ones that are not,
# and we keep track of the new length of the array with the slow pointer.
```

```python
# Two Pointers
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        left, right = 0, len(nums) - 1
        
        while left <= right:
            if nums[left] == val:
                # If the left pointer points to the target value,
                # we swap it with the right pointer's value.
                nums[left], nums[right] = nums[right], nums[left]
                # Then we decrement the right pointer.
                right -= 1
            else:
                # If the left pointer does not point to the target value,
                # we just increment it.
                left += 1
        
        return left

# The key point here is that we swap the elements when we find the target value at the left pointer,
# and we decrement the right pointer to effectively remove the target value from the array.
# This way, we ensure that all elements that are not equal to the target value are moved to the front of the array,
# and we return the new length of the array with the left pointer.
```

### [977. Squares of a Sorted Array](https://leetcode.com/problems/squares-of-a-sorted-array/)

```python
# two pointers, reverse order

class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:
        l, r = 0, len(nums) - 1
        ans = []

        while l <= r:
            if nums[l] ** 2 > nums[r] ** 2:
                ans.append(nums[l] ** 2)
                l += 1
            else:
                ans.append(nums[r] ** 2)
                r -= 1
        
        return ans[::-1]

# The key point here is that we compare the squares of the elements at the left and right pointers,
# and we append the larger square to the result list.
```

```python
# two pointers, append to the left using deque

from collections import deque
from typing import List

class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:
        l, r = 0, len(nums) - 1
        ans = deque() # use deque to append to the left

        while l <= r:
            if nums[l] ** 2 > nums[r] ** 2:
                ans.appendleft(nums[l] ** 2)
                l += 1
            else:
                ans.appendleft(nums[r] ** 2)
                r -= 1
        
        return list(ans) # convert deque to list

# The key point here is that we use a deque to append the squares of the elements at the left and right pointers to the left,
# ensuring that the result list is in sorted order without needing to reverse it at the end.
```
